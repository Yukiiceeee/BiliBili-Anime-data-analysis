```python
import pandas as pd
# 呈现图表效果导包，引入相关配置文件
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
# 读取数据表格
df = pd.read_excel('番剧数据-清洗后数据【2015-2023】.xlsx')
df.head()
```

```python
# 划分数据
total_play_volume_by_year = df.groupby('年份')['播放量'].sum()
# 获取年份和对应的播放量总和
year_info = total_play_volume_by_year.index.to_list()
volume_info = total_play_volume_by_year.tolist()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker

c = (
    Bar()
    .add_xaxis(year_info)
    .add_yaxis("播放量", volume_info)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="2015-2023年Bilibili番剧播放量"),
        datazoom_opts=opts.DataZoomOpts(),
    )
    .render("各年份番剧播放量信息饼状图.html")
)
```
